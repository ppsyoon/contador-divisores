```python
# Função de Entrada

def ler_numero():
  n = int(input("Digite um numero inteiro e positivo: "))
  while n <= 0:
    n = int(input("\nNumero invalido. Digite novamente: "))
  return n


# Função de cálculo

def contar_divisores(n):
  conta_divs = 0
  divisor = 1
  while divisor <= n:
    resto = n % divisor
    if resto == 0:
      conta_divs += 1
    divisor += 1
  return conta_divs

# Programa Principal

def main():
  n = ler_numero()
  qtd = contar_divisores(n)
  print(f"\nO número {n} tem {qtd} divisores")


# Chamada do Programa
main()
```
